```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
data = pd.read_excel('Cleaned_Data_With_Duplicates_ID.xlsx')
data.head(5)
```

```python
ID1= -1
year1 = 0
sport1 = "abcd"
m = 0
data2 = data.copy()
for rows in data2.iterrows():
    if ID1 == rows[1][0]:
        if year1 == rows[1][8]:
            if sport1 == rows[1][11]:
                data2.drop(m,inplace = True)
                m = m + 1
                continue
    ID1 = rows[1][0]
    year1 = rows[1][8]
    sport1 = rows[1][11]
    m = m + 1

data2.head(10)
```

```python
data2.to_excel('Cleaned_Data_Without_Duplicate_ID.xlsx', index = None, header=True)
```
